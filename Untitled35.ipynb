## Create New Features

### Subtask:
Derive new features that could be useful for modeling. This includes creating an 'Amenities_Count' by counting the number of amenities listed in the 'Amenities' column, and a 'Proximity_Score' by combining 'Nearby_Schools' and 'Nearby_Hospitals'.


**Reasoning**:
I will create a new column 'Amenities_Count' by counting amenities from the 'Amenities' column, handling missing values, and then create 'Proximity_Score' by summing 'Nearby_Schools' and 'Nearby_Hospitals', as instructed by the subtask.



````python
if 'Amenities' in df.columns:
    df['Amenities_Count'] = df['Amenities'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)
else:
    df['Amenities_Count'] = 0

df['Proximity_Score'] = df['Nearby_Schools'] + df['Nearby_Hospitals']

print("New features 'Amenities_Count' and 'Proximity_Score' created.")
print(df[['Amenities', 'Amenities_Count', 'Nearby_Schools', 'Nearby_Hospitals', 'Proximity_Score']].head())
````

## Encode Categorical Features

### Subtask:
Apply One-Hot Encoding to selected nominal categorical columns such as 'Property_Type', 'Furnished_Status', 'Public_Transport_Accessibility', 'Parking_Space', 'Security', 'Facing', 'Owner_Type', 'Availability_Status', 'State', and 'City' to convert them into a numerical format suitable for machine learning models. ## Scale Numerical Features

### Subtask:
Normalize or scale numerical features ('BHK', 'Size_in_SqFt', 'Price_in_Lakhs', 'Price_per_SqFt', 'Year_Built', 'Floor_No', 'Total_Floors', 'Age_of_Property', 'Nearby_Schools', 'Nearby_Hospitals', 'Amenities_Count', 'Proximity_Score') using StandardScaler to ensure that features with larger values do not dominate the learning process.


**Reasoning**:
First, I will define the list of numerical columns to be scaled. Then, I will filter this list to ensure that only columns present in the DataFrame and having a numerical data type are selected. After that, I will import and instantiate the `StandardScaler` to apply it to these selected numerical columns, transforming them to a standardized scale. Finally, I will display the head and info of the updated DataFrame to verify the changes.



````python
import numpy as np
from sklearn.preprocessing import StandardScaler

numerical_cols_to_scale = [
    'BHK', 'Size_in_SqFt', 'Price_in_Lakhs', 'Price_per_SqFt',
    'Year_Built', 'Floor_No', 'Total_Floors', 'Age_of_Property',
    'Nearby_Schools', 'Nearby_Hospitals', 'Amenities_Count', 'Proximity_Score'
]

# Ensure only existing numerical columns are selected for scaling
existing_numerical_cols = [col for col in numerical_cols_to_scale if col in df.columns and np.issubdtype(df[col].dtype, np.number)]

if existing_numerical_cols:
    scaler = StandardScaler()
    df[existing_numerical_cols] = scaler.fit_transform(df[existing_numerical_cols])
    print(f"\nScaled numerical features: {existing_numerical_cols}")
else:
    print("\nNo specified numerical features found for scaling.")

print("\n--- Preprocessed DataFrame Head after scaling ---")
print(df.head())

print("\n--- Preprocessed DataFrame Info after scaling ---")
df.info()
````

## Create Binary 'Good Investment' Label

### Subtask:
Calculate the 'Appreciation_Rate' based on 'Price_in_Lakhs' and 'Future_Price_5Y'. Then, create a binary target variable 'Good_Investment' by setting a threshold (e.g., 25th percentile of appreciation rate) to categorize properties as good or bad investments.


**Reasoning**:
First, I will calculate the 'Appreciation_Rate' using 'Future_Price_5Y' and 'Price_in_Lakhs'. Then, I will determine the 25th percentile of this rate to serve as the threshold for 'Good_Investment', and finally create the binary 'Good_Investment' column based on this threshold, printing the results to verify the new features.



````python
df['Appreciation_Rate'] = ((df['Future_Price_5Y'] - df['Price_in_Lakhs']) / df['Price_in_Lakhs']) * 100

threshold = df['Appreciation_Rate'].quantile(0.25)
df['Good_Investment'] = (df['Appreciation_Rate'] > threshold).astype(int)
print(f"Created 'Appreciation_Rate' and binary 'Good_Investment' label with threshold {threshold:.2f}%.")
print("\n--- Preprocessed DataFrame Head after creating investment label ---")
print(df[['Price_in_Lakhs', 'Future_Price_5Y', 'Appreciation_Rate', 'Good_Investment']].head())

print("\n--- Value counts for 'Good_Investment' ---")
print(df['Good_Investment'].value_counts())
````
